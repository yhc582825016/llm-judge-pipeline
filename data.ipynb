````python
import json
import ast

def _strip_code_fence(text: str) -> str:
    if text is None:
        return ""
    text = str(text)
    return text.replace("```json", "").replace("```", "").strip()

def _try_json_loads(s: str):
    # 直接按 JSON 解析
    return json.loads(s)

def _try_ast_literal_eval(s: str):
    # 1) 直接按 Python 字面量解析（常见于单引号、True/False/None）
    try:
        return ast.literal_eval(s)
    except Exception:
        pass

    # 2) 兼容一些“接近 JSON 但不规范”的情况：true/false/null
    #    将其转换成 Python 的 True/False/None 再 literal_eval
    s2 = (
        s.replace(": true", ": True")
         .replace(": false", ": False")
         .replace(": null", ": None")
    )
    return ast.literal_eval(s2)

def extract_score(text):
    s = _strip_code_fence(text)
    if not s:
        return None

    obj = None

    # 方法1：json.loads
    try:
        obj = _try_json_loads(s)
    except Exception:
        # 方法2：ast.literal_eval
        try:
            obj = _try_ast_literal_eval(s)
        except Exception:
            return None

    # 必须是 dict 且包含字段，否则返回 None
    if not isinstance(obj, dict):
        return None

    clarity = obj.get("Clarity", None)
    cog = obj.get("Cognitive_Complexity", None)

    # 两个字段任意一个缺失，也按你需求返回 None（你也可改成只缺一个就保留另一个）
    if clarity is None or cog is None:
        return None

    return {"Clarity": clarity['Score'], "Cognitive_Complexity": cog['Score']}

data["complex_clarity_rating"] = data["sglang_result"].apply(extract_score)
````

````python
import json
import pandas as pd
import numpy as np

def load_parquet(path):
    return pd.read_parquet(path)

def load_jsonl(path):
    with open(path,"r") as f:
        data = [json.loads(line) for line in f]
    return data
    
def load_json(path):
    with open(path, "r") as f:
        data = json.load(f)
    return data

def save_json(data, path):
    with open(path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

def save_jsonl(data, path):
    with open(path, 'w', encoding='utf-8') as f:
        for line in data:
            json_line = json.dumps(line, ensure_ascii=False)
            f.write(json_line + '\n')

def save_result(result, save_path):
    with open(save_path, 'a') as f:
        json.dump(result, f, ensure_ascii=False)
        f.write('\n')
        
def load_json_or_jsonl(file_path):
    try:
        with open(file_path, 'r') as f:
            return_data = json.load(f)
    except:
        with open(file_path, 'r') as f:
            data = f.readlines()
        return_data = []
        for i in range(len(data)):
            try:
                return_data.append(json.loads(data[i]))
            except:
                # print(data[i])
                continue
    return return_data
````
